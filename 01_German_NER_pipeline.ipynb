## Import a dataset of news headlines in German

```python
# Load this dataset: https://huggingface.co/datasets/Jotschi/german-news-titles
from datasets import load_dataset

ds = load_dataset("Jotschi/german-news-titles")
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)

df = pd.DataFrame(ds['train'])
df.head(5)
```

```python
titles_from_df = df['titles'].tolist()
print(titles_from_df[0])
```

```python
type(titles_from_df[0])
```

```python
# titles_from_df is a list of lists
# convert it to a list of strings

titles=[]

for sublist in titles_from_df:
  for item in sublist:
    titles.append(item)

print(titles[0])
```

```python
len(titles)
```

## Apply spaCy's German NER on titles

# Test the small German model

```python
!python -m spacy download de_core_news_sm
```

```python
# Process news titles with spaCy's German language model

import spacy

# Test model on one sentence
nlp_spacy_sm = spacy.load('de_core_news_sm')
doc_sentence = nlp_spacy_sm(titles[0])
print(doc_sentence.text)

# Check the entities
for token in doc_sentence:
  print(token.text, token.ent_type_)
```

```python
# Apply the model to the whole document
doc_spacy_sm = [nlp_spacy_sm(title) for title in titles]
```

```python
# Visualize a few examples:
for sentence in doc_spacy_sm[:3]:
  print(sentence.text)
  print("---")
  for ent in sentence.ents:
    print(ent.text, ent.label_)
  print("---------------------------------------------------------")
```

```python
# Save data as a csv file

import pandas as pd

rows=[]

for i, doc in enumerate(doc_spacy_sm):
  for ent in doc.ents:
    rows.append({
        'title': titles[i],
        'entity': ent.text,
        'label': ent.label_
                 })

df_spacy_sm = pd.DataFrame(rows)
df_spacy_sm.to_csv('df_spacy_sm.csv', index=False)
```

```python
df_spacy_sm.head(10)
```

## Apply stanza's German NER on titles

Refer to [Stanza-Beginners-Guide.ipynb](https://colab.research.google.com/github/stanfordnlp/stanza/blob/master/demo/Stanza_Beginners_Guide.ipynb#scrollTo=HDwRm-KXGcYo) and [Stanza - Named Entity Recognition](https://stanfordnlp.github.io/stanza/ner.html#accessing-named-entity-recogition-ner-tags-for-token)

```python
!pip install stanza
```

```python
import stanza

stanza.download('de')
```

```python
nlp_stanza = stanza.Pipeline('de', processors='tokenize, ner')
```

```python
doc_sentence_stanza = nlp_stanza(titles[0])

for sent in doc_sentence_stanza.sentences:
  print(sent.text)
  for ent in sent.ents:
    print(ent.text, ent.type)
```

```python
# Use the GPU as the CPU would take a long time to process

nlp_stanza_gpu = stanza.Pipeline('de', processors = 'tokenize,ner', use_gpu= True)
```

```python
# Use tqdm to keep track of processing progress

from tqdm import tqdm
doc_stanza = [nlp_stanza_gpu(title) for title in tqdm(titles)]
```

```python
# Visualize a few examples:
for sentence in doc_stanza[:3]:
  print(sentence.text)
  print("---")
  for ent in sentence.ents:
      print(ent.text,ent.type)
  print("---------------------------------------------------------")
```

```python
# Save data as a csv file
rows=[]

for i, doc in enumerate(doc_stanza):
  for ent in doc.ents:
    rows.append({
        'title': titles[i],
        'entity': ent.text,
        'label': ent.type
                 })

df_stanza = pd.DataFrame(rows)
df_stanza.to_csv('df_stanza.csv', index=False)
```
