## Import a dataset of news headlines in German

```python
# Load this dataset: https://huggingface.co/datasets/Jotschi/german-news-titles
from datasets import load_dataset

ds = load_dataset("Jotschi/german-news-titles")
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)

df = pd.DataFrame(ds['train'])
df.head(10)
```

```python
titles_from_df = df['titles'].tolist()
print(titles_from_df[0])
```

```python
type(titles_from_df[0])
```

```python
# titles_from_df is a list of lists
# convert it to a list of strings

titles=[]

for sublist in titles_from_df:
  for item in sublist:
    titles.append(item)

print(titles[0])
```

```python
len(titles)
```

## Apply spacy NER for German to titles

```python
!python -m spacy download de_core_news_sm
```

```python
# Process news titles with spaCy's German language model

import spacy

# Test model on one sentence
nlp = spacy.load('de_core_news_sm')
doc_sentence = nlp(titles[0])
print(doc_sentence.text)

# Check the entities
for token in doc_sentence:
  print(token.text, token.ent_type_)
```

```python
# Apply the model to the whole document
doc = [nlp(title) for title in titles]
```

```python
# Visualize a few examples:
for sentence in doc[:3]:
  print(sentence.text)
  print("---")
  for token in sentence:
    if token.ent_type_ != "":
      print(token.text, token.ent_type_)
  print("---------------------------------------------------------")

# The small German model is splitting Frankfurt am Main
```
